```python
# 6. Write a function to demonstrate the behaviour of mutable and immutable arguments.

def modify(mutable_list, immutable_tuple):
  mutable_list.append(4)

  try:
    immutable_tuple += (4,)


  except:
    pass

  return mutable_list, immutable_tuple


my_list = [1, 3, 7]
my_tuple = (1,2,3)

new_list, new_tuple = modify(my_list, my_tuple)
print("Original list after function:", my_list)
print("Original tuple after function:", my_tuple)
print("Return new tuple:", new_tuple)

```

```python
# 14. Reverse a string using while loop

s = input("Enter a string: ")

rev = ""
i = len(s) - 1

while i >= 0:
  rev += s[i]
  i -= 1

print("Reversed string:", rev)
```

```python
# 15. Factorial of a number using while loop

num = int(input("Enter a number:  "))
fact = 1

i = 1
while i <= num:
  fact *= i
  i += 1

print("Factorial =", fact)




```
